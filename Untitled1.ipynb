```python
# ============================================================
# STEP 1: IMPORT LIBRARIES & CREATE SPARK SESSION
# ============================================================

from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler, StringIndexer, StandardScaler, PCA
from pyspark.ml.classification import LogisticRegression, LogisticRegressionModel
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.sql.functions import col, when
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create SparkSession
spark = SparkSession.builder \
    .appName("CreditScorePrediction") \
    .config("spark.sql.adaptive.enabled", "true") \
    .getOrCreate()

print("✅ SparkSession created successfully!")
print(f"Spark Version: {spark.version}")
```

```python
# ============================================================
# STEP 2: LOAD DATASET
# ============================================================

df = spark.read.csv("credit_scoring_dataset.csv", header=True, inferSchema=True)

print(f"Total Records: {df.count()}")
print(f"Total Columns: {len(df.columns)}")
print("\nFirst 5 rows:")
df.show(5, truncate=False)

print("\nSchema:")
df.printSchema()
```

```python
# ============================================================
# STEP 1: DATA PREPARATION (35 Hrs)
# Objective: Clean, aggregate, and prepare data for model building
# ============================================================

print("=" * 60)
print("STEP 1: DATA PREPARATION")
print("=" * 60)

# 1.1 Check for missing values
print("\n1. Missing Values Check:")
for column in df.columns:
    null_count = df.filter(col(column).isNull()).count()
    if null_count > 0:
        print(f"  ⚠️  {column}: {null_count} missing values")
    else:
        print(f"  ✅ {column}: No missing values")

# 1.2 Feature Engineering - Create meaningful new features
print("\n2. Feature Engineering:")
df = df.withColumn("Loan_to_Income_Ratio", col("Loan_Amount") / col("Customer_Income"))
df = df.withColumn("Outstanding_Ratio", col("Outstanding") / col("Loan_Amount"))
df = df.withColumn("Price_per_Unit", col("Price") / col("Volume"))

print("  ✅ Loan_to_Income_Ratio = Loan_Amount / Customer_Income")
print("  ✅ Outstanding_Ratio = Outstanding / Loan_Amount")
print("  ✅ Price_per_Unit = Price / Volume")

# 1.3 Encode Categorical Variables using StringIndexer
print("\n3. Encoding Categorical Variables:")
categorical_cols = [
    "Customer_Gender", "Customer_Marital_Status", "Cust_State",
    "Product_Category", "Product_company", "Cust_Ethnic", "Occupation"
]

indexers = []
for col_name in categorical_cols:
    indexer = StringIndexer(
        inputCol=col_name,
        outputCol=f"{col_name}_indexed",
        handleInvalid="keep"
    )
    df = indexer.fit(df).transform(df)
    print(f"  ✅ {col_name} → {col_name}_indexed")

# 1.4 Select all features for the model
print("\n4. Feature Selection:")
numerical_features = [
    "Customer_Age", "Customer_Income", "Volume", "Price",
    "Loan_Amount", "Outstanding", "Loan_to_Income_Ratio",
    "Outstanding_Ratio", "Price_per_Unit"
]

indexed_cols = [f"{c}_indexed" for c in categorical_cols]
feature_cols = numerical_features + indexed_cols

print(f"  Total features selected: {len(feature_cols)}")
print(f"  Numerical: {len(numerical_features)}")
print(f"  Categorical (encoded): {len(indexed_cols)}")

# 1.5 Assemble features into single vector using VectorAssembler
print("\n5. VectorAssembler - Combining features into single column:")
assembler = VectorAssembler(
    inputCols=feature_cols,
    outputCol="raw_features",
    handleInvalid="skip"
)
df = assembler.transform(df)
print("  ✅ Features assembled into 'raw_features' column")

# 1.6 Standardize features using StandardScaler
print("\n6. StandardScaler - Normalizing features:")
scaler = StandardScaler(
    inputCol="raw_features",
    outputCol="features",
    withStd=True,
    withMean=True
)
scaler_model = scaler.fit(df)
df = scaler_model.transform(df)
print("  ✅ Features standardized into 'features' column")

# 1.7 PCA for dimensionality reduction (reducing correlated variables)
print("\n7. PCA - Dimensionality Reduction:")
pca = PCA(k=10, inputCol="features", outputCol="pca_features")
pca_model = pca.fit(df)
df = pca_model.transform(df)
explained_variance = pca_model.explainedVariance.toArray()
print(f"  ✅ PCA applied (10 components)")
print(f"  Explained Variance: {explained_variance.sum():.4f}")

# 1.8 Prepare final dataset with label
print("\n8. Preparing final dataset:")
final_data = df.select("features", "Default_Risk").withColumnRenamed("Default_Risk", "label")
print("  ✅ Label column: 'label' (0=No Risk, 1=Risk)")

# 1.9 Split data: 75% Training, 25% Testing
print("\n9. Train-Test Split:")
train_data, test_data = final_data.randomSplit([0.75, 0.25], seed=42)
print(f"  ✅ Training set: {train_data.count()} records")
print(f"  ✅ Testing set: {test_data.count()} records")

# Check class distribution
print("\n10. Class Distribution in Training Set:")
train_data.groupBy("label").count().show()

print("\n" + "=" * 60)
print("STEP 1 COMPLETED: Data Preparation Done!")
print("=" * 60)
```

```python
# ============================================================
# STEP 2: BUILD MODEL (35 Hrs)
# Objective: Build logistic regression model to predict credit default risk
# ============================================================

print("=" * 60)
print("STEP 2: BUILD MODEL")
print("=" * 60)

# 2.1 Create LogisticRegression instance
print("\n1. Creating Logistic Regression Model:")
lr = LogisticRegression(
    featuresCol="features",
    labelCol="label",
    maxIter=100,
    regParam=0.01,        # Regularization parameter
    elasticNetParam=0.0,   # 0 = L2 regularization
    family="binomial",     # Binary classification
    threshold=0.5
)
print("  ✅ Model configured:")
print("     - featuresCol: 'features'")
print("     - labelCol: 'label'")
print("     - maxIter: 100")
print("     - regParam: 0.01")
print("     - elasticNetParam: 0.0 (L2)")
print("     - family: binomial")

# 2.2 Fit the model on training data
print("\n2. Training the model...")
model = lr.fit(train_data)
print("  ✅ Model trained successfully!")

# 2.3 Examine model coefficients and intercept
print("\n3. Model Parameters Analysis:")
print("-" * 50)

print(f"\n  Intercept (Bias): {model.intercept:.6f}")
print(f"\n  Coefficients (Weights):")
coefficients = model.coefficients.toArray()

# Map coefficients to feature names
all_feature_names = numerical_features + categorical_cols
print(f"\n  {'Feature':<25} {'Coefficient':>12} {'Impact'}")
print(f"  {'-'*25} {'-'*12} {'-'*20}")

for i, (name, coef) in enumerate(zip(all_feature_names[:len(coefficients)], coefficients)):
    direction = "↑ Increases Risk" if coef > 0 else "↓ Decreases Risk"
    print(f"  {name:<25} {coef:>12.6f}  {direction}")

# 2.4 Feature Importance Analysis
print("\n4. Feature Importance Ranking:")
print("-" * 50)
print("(Higher absolute coefficient = Greater impact on credit score)\n")

feature_importance = list(zip(all_feature_names[:len(coefficients)], coefficients))
feature_importance.sort(key=lambda x: abs(x[1]), reverse=True)

print(f"  {'Rank':<6} {'Feature':<25} {'Coefficient':>12} {'Direction'}")
print(f"  {'-'*6} {'-'*25} {'-'*12} {'-'*15}")

for i, (name, coef) in enumerate(feature_importance[:10], 1):
    direction = "INCREASES" if coef > 0 else "DECREASES"
    print(f"  {i:<6} {name:<25} {coef:>12.6f}  {direction}")

# 2.5 Training Summary
print("\n5. Training Summary:")
print("-" * 50)

training_summary = model.summary
print(f"  Training AUC-ROC:  {training_summary.areaUnderROC:.4f}")
print(f"  Training Accuracy: {training_summary.accuracy:.4f}")
print(f"  Training Precision: {training_summary.precisionByLabel[1]:.4f}")
print(f"  Training Recall:   {training_summary.recallByLabel[1]:.4f}")
print(f"  Training F1-Score: {training_summary.fMeasureByLabel()[1]:.4f}")

print("\n" + "=" * 60)
print("STEP 2 COMPLETED: Model Built Successfully!")
print("=" * 60)
```

```python
# ============================================================
# STEP 3: EVALUATION AND TUNE (40 Hrs)
# Objective: Evaluate model, find best hyperparameters
# ============================================================

print("=" * 60)
print("STEP 3: EVALUATION AND TUNE")
print("=" * 60)

# 3.1 Make predictions on test dataset
print("\n1. Making predictions on test data...")
predictions = model.transform(test_data)

print("\n   Sample Predictions:")
predictions.select("label", "prediction", "probability").show(10, truncate=False)

# 3.2 Evaluate using AUC-ROC (PRIMARY METRIC)
print("\n2. Model Evaluation Metrics:")
print("-" * 50)

roc_evaluator = BinaryClassificationEvaluator(
    labelCol="label",
    rawPredictionCol="rawPrediction",
    metricName="areaUnderROC"
)
auc_roc = roc_evaluator.evaluate(predictions)
print(f"   AUC-ROC (Primary Metric): {auc_roc:.4f}")

# AUC-PR
pr_evaluator = BinaryClassificationEvaluator(
    labelCol="label",
    rawPredictionCol="rawPrediction",
    metricName="areaUnderPR"
)
auc_pr = pr_evaluator.evaluate(predictions)
print(f"   AUC-PR:                   {auc_pr:.4f}")

# Accuracy
accuracy_evaluator = MulticlassClassificationEvaluator(
    labelCol="label", predictionCol="prediction", metricName="accuracy"
)
accuracy = accuracy_evaluator.evaluate(predictions)
print(f"   Accuracy:                 {accuracy:.4f}")

# Precision
precision_evaluator = MulticlassClassificationEvaluator(
    labelCol="label", predictionCol="prediction", metricName="weightedPrecision"
)
precision = precision_evaluator.evaluate(predictions)
print(f"   Precision:                {precision:.4f}")

# Recall
recall_evaluator = MulticlassClassificationEvaluator(
    labelCol="label", predictionCol="prediction", metricName="weightedRecall"
)
recall = recall_evaluator.evaluate(predictions)
print(f"   Recall:                   {recall:.4f}")

# F1 Score
f1_evaluator = MulticlassClassificationEvaluator(
    labelCol="label", predictionCol="prediction", metricName="f1"
)
f1_score = f1_evaluator.evaluate(predictions)
print(f"   F1-Score:                 {f1_score:.4f}")

# 3.3 Confusion Matrix
print("\n3. Confusion Matrix:")
print("-" * 50)

# Convert to pandas for confusion matrix
predictions_pd = predictions.select("label", "prediction").toPandas()
tp = predictions_pd[(predictions_pd["label"] == 1) & (predictions_pd["prediction"] == 1)].shape[0]
tn = predictions_pd[(predictions_pd["label"] == 0) & (predictions_pd["prediction"] == 0)].shape[0]
fp = predictions_pd[(predictions_pd["label"] == 0) & (predictions_pd["prediction"] == 1)].shape[0]
fn = predictions_pd[(predictions_pd["label"] == 1) & (predictions_pd["prediction"] == 0)].shape[0]

print(f"\n                    Predicted")
print(f"                 No Risk   Risk")
print(f"   Actual No Risk:   {tn:3d}     {fp:3d}")
print(f"   Actual Risk:      {fn:3d}     {tp:3d}")

print(f"\n   True Negatives (TN):  {tn}")
print(f"   False Positives (FP): {fp}")
print(f"   False Negatives (FN): {fn}")
print(f"   True Positives (TP):  {tp}")

# 3.4 Fine Tuning with Cross-Validation
print("\n4. Hyperparameter Tuning (Cross-Validation):")
print("-" * 50)
print("   Setting up ParamGrid with 5-Fold CV...")
print("   (This may take 1-2 minutes)\n")

param_grid = ParamGridBuilder() \
    .addGrid(lr.regParam, [0.001, 0.01, 0.1]) \
    .addGrid(lr.elasticNetParam, [0.0, 0.5]) \
    .addGrid(lr.maxIter, [50, 100]) \
    .build()

cv = CrossValidator(
    estimator=lr,
    estimatorParamMaps=param_grid,
    evaluator=BinaryClassificationEvaluator(metricName="areaUnderROC"),
    numFolds=5,
    seed=42
)

cv_model = cv.fit(train_data)
best_model = cv_model.bestModel

print("   ✅ Cross-Validation Complete!")
print(f"\n   Best Model Parameters:")
print(f"   - regParam:        {best_model._java_obj.getRegParam()}")
print(f"   - elasticNetParam: {best_model._java_obj.getElasticNetParam()}")
print(f"   - maxIter:         {best_model._java_obj.getMaxIter()}")

# Evaluate best model on test data
best_predictions = best_model.transform(test_data)
best_auc_roc = roc_evaluator.evaluate(best_predictions)
print(f"\n   Best Model AUC-ROC on Test: {best_auc_roc:.4f}")
print(f"   Initial Model AUC-ROC:      {auc_roc:.4f}")
print(f"   Improvement:                {best_auc_roc - auc_roc:.4f}")

# Use best model going forward
model = best_model
predictions = best_predictions

print("\n" + "=" * 60)
print("STEP 3 COMPLETED: Evaluation and Tuning Done!")
print("=" * 60)
```

```python
# ============================================================
# STEP 4: DEPLOYMENT (25 Hrs)
# Objective: Save, load model and plot results for inference
# ============================================================

print("=" * 60)
print("STEP 4: DEPLOYMENT")
print("=" * 60)

# 4.1 Save the model
print("\n1. Saving the trained model...")
model_path = "credit_score_logistic_model"
model.write().overwrite().save(model_path)
print(f"   ✅ Model saved to: {model_path}")

# 4.2 Load the model
print("\n2. Loading the saved model...")
loaded_model = LogisticRegressionModel.load(model_path)
print("   ✅ Model loaded successfully!")

# Verify loaded model works
loaded_predictions = loaded_model.transform(test_data)
loaded_auc = roc_evaluator.evaluate(loaded_predictions)
print(f"   Loaded Model AUC-ROC: {loaded_auc:.4f}")

# 4.3 Generate Visualization Plots
print("\n3. Generating visualization plots...")

# Convert predictions to pandas for plotting
pred_pd = predictions.select("label", "prediction", "probability").toPandas()
pred_pd["prob_default"] = pred_pd["probability"].apply(lambda x: x[1])

# Plot 1: Confusion Matrix Heatmap
print("   📊 Plot 1: Confusion Matrix...")
cm = [[tn, fp], [fn, tp]]
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Default Risk (0)', 'Default Risk (1)'],
            yticklabels=['No Default Risk (0)', 'Default Risk (1)'],
            linewidths=2, linecolor='white', annot_kws={"size": 14})
plt.title('Confusion Matrix - Credit Default Prediction\n(Logistic Regression Model)',
          fontsize=14, fontweight='bold')
plt.ylabel('Actual Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.tight_layout()
plt.savefig("confusion_matrix.png", dpi=300, bbox_inches="tight")
plt.show()

# Plot 2: ROC Curve
print("   📊 Plot 2: ROC Curve...")
from sklearn.metrics import roc_curve, auc
fpr, tpr, _ = roc_curve(pred_pd["label"], pred_pd["prob_default"])
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=3, label=f'ROC Curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier (AUC = 0.50)')
plt.fill_between(fpr, tpr, alpha=0.2, color='darkorange')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=12)
plt.ylabel('True Positive Rate (Sensitivity / Recall)', fontsize=12)
plt.title('ROC Curve - Credit Default Prediction\n(Logistic Regression Model)', fontsize=14, fontweight='bold')
plt.legend(loc="lower right", fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("roc_curve.png", dpi=300, bbox_inches="tight")
plt.show()

# Plot 3: Feature Importance
print("   📊 Plot 3: Feature Importance...")
coefficients = model.coefficients.toArray()
all_feature_names = numerical_features + categorical_cols

coef_df = pd.DataFrame({
    'Feature': all_feature_names[:len(coefficients)],
    'Coefficient': coefficients
})
coef_df['Abs_Coefficient'] = coef_df['Coefficient'].abs()
coef_df = coef_df.sort_values('Abs_Coefficient', ascending=True)

plt.figure(figsize=(10, 8))
colors = ['#2ecc71' if c > 0 else '#e74c3c' for c in coef_df['Coefficient']]
plt.barh(coef_df['Feature'], coef_df['Coefficient'], color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)
plt.axvline(x=0, color='black', linestyle='-', linewidth=1.2)
plt.xlabel('Coefficient Value (Log-Odds Impact)', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.title('Feature Importance - Logistic Regression Coefficients\n(Higher Absolute Value = Greater Impact on Credit Score)',
          fontsize=13, fontweight='bold')
plt.grid(True, alpha=0.3, axis='x')

from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#2ecc71', label='Increases Default Risk'),
                   Patch(facecolor='#e74c3c', label='Decreases Default Risk')]
plt.legend(handles=legend_elements, loc='lower right', fontsize=10)
plt.tight_layout()
plt.savefig("feature_importance.png", dpi=300, bbox_inches="tight")
plt.show()

# Plot 4: Model Metrics Bar Chart
print("   📊 Plot 4: Model Metrics...")
metrics_names = ['AUC-ROC', 'Accuracy', 'Precision', 'Recall', 'F1-Score']
metrics_values = [loaded_auc, accuracy, precision, recall, f1_score]
colors_metrics = ['#3498db', '#2ecc71', '#9b59b6', '#e67e22', '#1abc9c']

plt.figure(figsize=(10, 6))
bars = plt.bar(metrics_names, metrics_values, color=colors_metrics, edgecolor='black', linewidth=1.2, alpha=0.85)
plt.ylim(0, 1.1)
plt.ylabel('Score', fontsize=12)
plt.title('Model Evaluation Metrics - Credit Default Prediction\n(Logistic Regression on Test Set)',
          fontsize=14, fontweight='bold')
plt.axhline(y=0.9, color='red', linestyle='--', linewidth=1.5, alpha=0.7, label='Target Threshold (0.90)')

for bar, val in zip(bars, metrics_values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
             f'{val:.4f}', ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.legend(fontsize=11)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig("model_metrics.png", dpi=300, bbox_inches="tight")
plt.show()

# 4.4 Inference on New Data
print("\n4. Inference on Sample Customer Data:")
print("-" * 50)

# Create a sample inference record
sample_data = spark.createDataFrame([
    (35, 500000, 1, 800000, 600000, 300000, 1.2, 0.5, 800000,  # numerical
     0, 0, 3, 2, 1, 0, 2, 0)  # categorical indexed values
], numerical_features + indexed_cols)

sample_assembled = assembler.transform(sample_data)
sample_scaled = scaler_model.transform(sample_assembled)
sample_pred = loaded_model.transform(sample_scaled)
sample_result = sample_pred.select("prediction", "probability").collect()[0]

predicted_class = int(sample_result["prediction"])
prob_risk = sample_result["probability"][1]
prob_safe = sample_result["probability"][0]

print(f"\n   Sample Customer Profile:")
print(f"   - Age: 35 years")
print(f"   - Income: ₹5,00,000")
print(f"   - Loan Amount: ₹6,00,000")
print(f"   - Outstanding: ₹3,00,000")
print(f"   - Outstanding Ratio: 50%")

print(f"\n   🔮 Prediction Result:")
print(f"   - Predicted Class: {'⚠️ DEFAULT RISK' if predicted_class == 1 else '✅ NO RISK'}")
print(f"   - Probability of Default: {prob_risk:.4f} ({prob_risk*100:.2f}%)")
print(f"   - Probability of Safe:    {prob_safe:.4f} ({prob_safe*100:.2f}%)")

print("\n" + "=" * 60)
print("STEP 4 COMPLETED: Deployment Done!")
print("=" * 60)

print("\n" + "=" * 60)
print("🎉 PROJECT COMPLETED SUCCESSFULLY!")
print("=" * 60)
print(f"""
Final Model Performance:
  AUC-ROC (Primary):  {loaded_auc:.4f}
  Accuracy:           {accuracy:.4f}
  Precision:          {precision:.4f}
  Recall:             {recall:.4f}
  F1-Score:           {f1_score:.4f}

Files Generated:
  ✅ confusion_matrix.png
  ✅ roc_curve.png
  ✅ feature_importance.png
  ✅ model_metrics.png
  ✅ credit_score_logistic_model/ (saved model)

Next Steps:
  1. Download all plots from the Files panel (left sidebar)
  2. Upload code to GitHub/Google Drive
  3. Create PPT/PDF with plots
  4. Record execution video
  5. Submit on TCS iON portal
""")
```
